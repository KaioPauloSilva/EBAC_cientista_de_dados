````python
# Imports
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf


import scipy.stats as ss 
from statsmodels.stats.outliers_influence import variance_inflation_factor as vif

import patsy
import warnings

import statsmodels.api as sm

from scipy.interpolate import interp1d
sns.set_theme()
````

````python
df = pd.read_csv('./previsao_de_renda_II.csv', index_col=0)
df.head()
````

````python
df.info()
````

````python
# Preenche os valores missing pela mdia
df['tempo_emprego'].fillna(value=df['tempo_emprego'].mean(), inplace=True)
````

````python
# Separando bases de treino e teste
df_train = df.loc[:'2015-12-01'].copy()
df_test = df.loc['2016-01-01':].copy()

# Criando as variveis log_renda
df_train['log_renda'] = np.log(df_train['renda'])
df_test['log_renda'] = np.log(df_test['renda'])
````

## Verifique as suposies do modelo

Faa uma anlise para avaliar as suposies bsicas:

- Os resduos desse modelo possuem distribuio Normal (ou algo prximo)?
- Verifique a independncia dos resduos
    - Faa um grfico dos resduos versus os valores preditos
    - Avalie se h padres dos resduos versus cada uma das variveis do modelo
    - Avalie se o valor mdio dos resduos aparenta ter relao com o ms de referncia
- Avalie se a varincia da varivel resposta parece ser conforme os valores previstos.

Voc considera que as suposies esto atendidas? H algum impacto em eventuais desvios?

````python
reg1 = patsy.dmatrices('''log_renda ~ C(posse_de_imovel)
                    + qtd_filhos 
                    + idade
                    + tempo_emprego
                    + 1''', df_train)

res = smf.ols(reg1, data=df_train).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 1
                         , alpha = 0.05)
````

````python
res.summary()
````

````python
# Vamos remover mais uma varivel do modelo j que o R no se alterou
# A varivel idade ser removida, pois representa um coeficiente muito pequeno

reg1 = patsy.dmatrices('''log_renda ~ C(posse_de_imovel)
                    + tempo_emprego
                    + 1''', df_train)

res = smf.ols(reg1, data=df_train).fit_regularized(method = 'elastic_net' 
                         , refit = True
                         , L1_wt = 1
                         , alpha = 0.05)
````

````python
res.summary()
````

````python
sns.displot(res.resid, height=3, aspect=2)
plt.show()
````

````python
sns.residplot(x=res.predict(), y='log_renda', data=df_train)
plt.show()
````

````python
sns.scatterplot(x = res.predict(), y = 'posse_de_imovel', data=df_train)
plt.show()
````

````python
sns.scatterplot(x = res.predict(), y = 'tempo_emprego', data=df_train);
plt.show()
````

````python
sns.scatterplot(x = res.predict(), y = 'data_ref', data=df_train);
plt.show()
````

````python
sns.scatterplot(x = res.predict(), y = res.resid)
plt.show()
````
