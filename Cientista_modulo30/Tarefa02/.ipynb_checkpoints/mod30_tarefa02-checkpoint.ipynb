```python
df.columns
```

```python
variaveis = ['Administrative', 'Administrative_Duration', 'Informational',
       'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration','Revenue','SpecialDay','Month','Weekend']

variaveis_quant = ['Administrative', 'Administrative_Duration', 'Informational',
       'Informational_Duration', 'ProductRelated', 'ProductRelated_Duration']

variaveis_quali = ['SpecialDay','Month','Weekend']

df_pad = pd.DataFrame(StandardScaler().fit_transform(df[variaveis_quant]), columns=df[variaveis_quant].columns)
df_pad[variaveis_quali] = df[variaveis_quali]
df_pad
```

```python
df_trat = pd.get_dummies(df_pad.dropna(), columns = variaveis_quali)
df_trat
```

## Nmero de grupos

Nesta atividade vamos adotar uma abordagem bem pragmtica e avaliar agrupamentos hierrquicos com 3 e 4 grupos, por estarem bem alinhados com uma expectativa e estratgia do diretor da empresa. 

*Ateno*: Cuidado se quiser fazer o dendrograma, pois com muitas observaes ele pode ser mais complicado de fazer, e dependendo de como for o comando, ele pode travar o *kernell* do seu python.

```python
df_trat.columns
```

```python
vars_cat = [True if x in {'SpecialDay_0.0', 'SpecialDay_0.2', 'SpecialDay_0.4', 'SpecialDay_0.6',
       'SpecialDay_0.8', 'SpecialDay_1.0', 'Month_Aug', 'Month_Dec',
       'Month_Feb', 'Month_Jul', 'Month_June', 'Month_Mar', 'Month_May',
       'Month_Nov', 'Month_Oct', 'Month_Sep', 'Weekend_False', 'Weekend_True'} else False for x in df_trat.columns]
vars_cat
```

```python
%%time
distancia_gower = gower_matrix(df_trat, cat_features=vars_cat)

gdv = squareform(distancia_gower, force='tovector')
```

```python
Z = linkage(gdv, method='complete')
```

### Anlise 3 Grupos :

```python
df_trat['grupo_3'] = fcluster(Z, 3, criterion='maxclust')
df_trat.grupo_3.value_counts()
```

```python
pd.crosstab(df_trat['grupo_3'], df.Revenue).transpose()
```

### Anlise 4 Grupos :

```python
df_trat['grupo_4'] = fcluster(Z, 4, criterion='maxclust')
df_trat.grupo_4.value_counts()
```

```python
pd.crosstab(df_trat['grupo_4'], df.Revenue).transpose()
```
